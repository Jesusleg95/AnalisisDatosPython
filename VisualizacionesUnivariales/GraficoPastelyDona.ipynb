```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df
```

**Objetivo Actual:** Visualizar la proporcion de pasajeros hombres y mujeres en el Titanic

```python
pasajeros = df.groupby('Sex').agg({
    'PassengerId':'count'
})
pasajeros
```

**Grafico de pastel en pandas**

```python
pasajeros.plot.pie(
    y="PassengerId"
)
```

```python
pasajeros.plot.pie(
    y="PassengerId",
    startangle=90
)
```

```python
pasajeros.plot.pie(
    y="PassengerId",
    startangle=90,
    autopct='%1.1f%%'
)
```

**Grafico de pastel con plotly**

```python
import plotly.express as px
```

```python
pasajeros2 = pasajeros.reset_index()
pasajeros2
```

```python
fig = px.pie(
    pasajeros2,
    names='Sex',
    values='PassengerId',
    title='Proporcion de pasajeros del Titanic por sexo'
)

fig.show()
```

**Grafico de Dona**

```python
fig2 = px.pie(
    pasajeros2,
    names='Sex',
    values='PassengerId',
    title='Proporcion de pasajeros del Titanic por sexo',
    hole=0.5
)

fig2.show()
```
