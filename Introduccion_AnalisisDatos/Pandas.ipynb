```python
import pandas as pd
import random
```

index_col es un parametro que nos ayuda a acomodar los datos por la columna que queramos

```python

df = pd.read_csv("dataset.csv", index_col="id")
df.head()
```

describe() devuelve un nuevo dataframe con informacion estadistica del dataframe original

```python
df.describe()
```

## **Limpieza de datos**

```python
df.head()
```

dropna() elimina las filas que tienen un valor "NaN" en alguna de sus columnas

```python
df_filtrador = df.dropna()
df_filtrador.head()
```

fillna() cambia el valor "NaN" por el valor pasado como parametro

```python
df_filtrado = df.fillna(0)
df_filtrado.head()
```

En la funcion fillna() podemos pasar un diccionario como parametro, de esta manera cambiar a diferentes valores dependiendo de la columna

```python
df = df.fillna({"retweets":0,"mentions":1})
df.head()
```

## **Filtrado de datos**

**Filtrar por Columna**

Para filtrar por una columna se agrega un par de corchetes, para filtrar por 2 o mas columnas se agregan dos pares de corchetes

```python
df["user"]
```

```python
df[["favorites", "full_text"]]
```

**Filtrar por fila**

con la funcion iloc[] se filtra la fila por el indice puesto por el notebook

```python
df.iloc[0]
```

```python
df.iloc[0:3]
```

```python
df.iloc[[0, 1, 2]]
```

La funcion loc permite filtrar las filas por el indice

```python
df.loc[[183721, 183722]]
```

**Filtrar por Filas y Columnas**

Para filtrar con filas y columnas se utiliza la funcion loc: **loc[*filas*][*columnas*]**

```python
df.loc[[183721, 183722], ["favorites"]]
```

**Filtrar por Condiciones**

```python
df[df["favorites"] > 400]
```

```python
df[(df["favorites"] > 400) & (df["mentions"] > 20)]
```

**Filtrado por Condicion en texto**

```python
df[df["full_text"].str.contains("Programming")]
```

## **Transformación de Datos**

```python
def calcularGanancias(retweets):
  ganancia = retweets * random.randint(3, 5)
  return ganancia

df["ganancias"] = df["retweets"].apply(calcularGanancias)

def popularidad(fila):
  resultado = fila["followees"]/fila["followers"]
  return resultado

#Dado que estamos haciendo .apply() del DataFrame completo y no de una columna individual, "axis = 1" le especifica a .apply() que la funcion debe ser aplicada por cada fila del DataFrame
df["popularidad"] = df.apply(popularidad, axis=1)

df.head()
```

## **Agrupación de Datos**

```python
df[['country', 'favorites', 'retweets', 'mentions', 'followers', 'followees', 'ganancias', 'popularidad']].groupby("country").mean()
```

```python
df.groupby("country").agg({
    "followers":'sum',
    "mentions":'mean',
    "retweets":'max'
})
```

**Filtrado de Agrupación de Datos**

```python
grouped = df.groupby("country").agg({
    "followers":'sum',
    "mentions":'mean',
    "retweets":'max'
})

grouped[grouped["followers"] > 5000]
```

## **Guardado de Datos**

```python
grouped.to_csv("salida.csv")
```
