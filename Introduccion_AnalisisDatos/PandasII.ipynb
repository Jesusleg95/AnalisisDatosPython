```python
import pandas as pd
```

```python
df1 = pd.read_csv("dataset_2a.csv")
df2 = pd.read_csv("dataset_2b.csv")
df3 = pd.read_csv("dataset_2c.csv")
```

## **JOINS**

La funcion merge() es el equivalente a la funcion join() de SQL

```python
df_con_usuarios = df1.merge(df2, on="user", how="left")
df_con_usuarios
```

```python
df_completo = df_con_usuarios.merge(df3, on="ciudad", how="left")
df_completo
```

## **Fechas**

```python
df_completo["fecha"] = pd.to_datetime(df_completo["fecha_publicado"], format="%d/%m/%Y")
df_completo
```

```python
df_completo["mes"] = df_completo["fecha"].apply(lambda x: x.strftime("%m/%Y"))
df_completo
```

```python
df_completo["dia"] = df_completo["fecha"].apply(lambda x: x.strftime("%A"))
df_completo
```

```python
df_completo["date"] = df_completo["fecha_publicado"].apply(lambda x: x[3:])
df_completo
```

## **Agrupacion**

```python
df_completo.groupby("mes").agg({
    "id":'count'
})
```

```python
agrupado = df_completo.groupby(["mes","ciudad"]).agg({
    "id":'count'
})

agrupado
```

```python
desagrupado = agrupado.reset_index(1)
desagrupado
```

## **Pivotar**
Convertir cada valor √∫nico de una columna en nuevas columnas

```python
pivote = desagrupado.pivot(columns="ciudad")
pivote
```

```python
pivote = pivote.droplevel(0, axis="columns")
pivote
```

```python
pivote[["GUAYAQUIL"]]
```

```python
pivote = pivote.fillna(0)
pivote
```

## **Transponer**
Convertir las columnas en filas y las filas en columnas

```python
pivote.transpose()
```
